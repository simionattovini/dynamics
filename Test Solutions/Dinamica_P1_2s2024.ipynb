### a) Para calcular a velocidade do paraquedista, é necessário definir um sistema de referência inercial e escrever a posição do paraquedista em função das variáveis conhecidas.

Assim, define-se os vetores:

$$
\vec{r}_b = \left\{
  \begin{array}{c}
    p_x \\ 0 \\ 0
  \end{array}
\right\}
\hspace{5mm},\hspace{5mm}
\vec{r}_p = \left\{
\begin{array}{c}
-r\cos\theta \\ r\sin \theta \\ 0
\end{array}
\right\}
$$

Neste caso, $p_x$ é a posição da lancha.

A posição do paraquedista é dada por:

$$
\vec{r} = \vec{r}_b + \vec{r}_p = \left\{
\begin{array}{c}
p_x - r\cos\theta \\ r\sin\theta \\ 0
\end{array}
\right\}
$$

Assim sua velocidade é dada por:

$$
\vec{v} = \left\{
  \begin{array}{c}
    v_x + r\dot{\theta}\sin\theta \\ r\dot{\theta}\cos\theta \\ 0
  \end{array}
\right\}
$$

Assim, $v_x$ é a velocidade da lancha em $x$.

Do enunciado, $v_x = 30\,\mathrm{km/h} \approx 8,333\,\mathrm{m/s}$ e $\dot\theta = 2^\circ/\mathrm{s} = \pi/90\,\mathrm{rad/s}$. Assim, a velocidade do paraquedista é de:

$$
\vec{v} = \left\{
  \begin{array}{c}
      11,824 \\ 6,046 \\ 0
  \end{array}
\right\} \,\mathrm{m/s}
$$


```python
from numpy import pi
import numpy as np

r = 200           # m
theta = 30*pi/180 # rad
vx = 30/3.6       # km/h -> m/s

dtheta = pi/90 # rad/s

v = np.asarray([vx + r*dtheta*np.sin(theta), r*dtheta*np.cos(theta), 0])

print(f'v = [{v[0]:.4f}, {v[1]:.4f}, {v[2]:.4f} ]ˆT m/s')
```

### b)

A aceleração do paraquedista é dada por:

$$
\vec{a} = \left\{
  \begin{array}{c}
    a_x + r\ddot{\theta}\sin\theta +r\dot{\theta}^2\cos\theta \\
    r\ddot{\theta}\cos\theta - r \dot{\theta}^2\sin\theta \\
    0
  \end{array}
\right\}
$$

Segundo o enunciado, a velocidade angular do cabo e a velocidade da lancha são constantes. Assim:

$$
\ddot{\theta} = a_x = 0
$$

O que simplifica a aceleração para:

$$
\vec{a} = \left\{
  \begin{array}{c}
      r\dot{\theta}^2\cos\theta \\
      - r \dot{\theta}^2\sin\theta \\
      0
  \end{array}
  \right\}
$$

Do enunciado, $r = 200\,\mathrm{m}$ e $\dot{\theta} = 2^\circ/\mathrm{s} = \pi/90 \,\mathrm{rad/s}$.

O que rende a aceleração de:

$$
\vec{a} = \left\{
  \begin{array}{c}
      0,211 \\
      -0,1218 \\
      0
  \end{array}
  \right\} \,\mathrm{m/s^2}
$$


```python
a = np.asarray([r*dtheta**2*np.cos(theta), -r*dtheta**2*np.sin(theta), 0])

print(f'a = [{a[0]:.4f}, {a[1]:.4f}, {a[2]:.4f} ]ˆT rad/s²')
```

### c)

Do enunciado, sabe-se que a massa do paraquedista é de 80 kg, a aceleração da gravidade é de 9,81 m/s² e a tração no cabo é de 2,4 kN.

Assim, pode-se aplicar a segunda lei de newton na forma:

$$
m\vec{a} = \vec{W} + \vec{T} + \vec{F}
$$

Neste caso, a aceleração é a encontrada anteriormente e os vetores $\vec{W}$ e $\vec{T}$ são, respectivamente:

$$
\vec{W} = \left\{\begin{array}{c}
0 \\ -mg \\ 0
\end{array}\right\}
= \left\{\begin{array}{c}
0 \\ -784,8 \\ 0
\end{array}\right\} \,\mathrm N
$$

$$
\vec{T} = \left\{\begin{array}{c}
2.400\cos\theta \\ -2400\sin\theta \\ 0
\end{array}\right\}
= \left\{\begin{array}{c}
2.078,46 \\ 1.200,00 \\ 0
\end{array}\right\} \,\mathrm N
$$

Assim:

$$
\vec{F} = m\vec{a}-\vec{W}-\vec{T} =
\left\{\begin{array}{c}
-2.061,58 \\ 1.975,05 \\ 0
\end{array}\right\} \,\mathrm N
$$

Este vetor representa uma força de 2.854,98 N com um ângulo de 136,23° com a horizontal.

```python
m = 80 # kg
g = 9.81 # m/s²
t = 2.4e3 # N

W = np.asarray([0, -m*g, 0])
T = np.asarray([t*np.cos(theta), -t*np.sin(theta), 0])

F = m*a - W - T

print(f'F = [{F[0]:.2f}, {F[1]:.2f}, {F[2]:.2f} ]ˆT N')
print(f'|F| = {np.linalg.norm(F):.2f} N')
print(f'Ângulo: {np.arctan2(F[1],F[0])*180/pi:.2f}°')
```

### Resolução

Na iminência do empinamento a empilhadeira deverá pivotar em torno do ponto $H$, fazendo com que a força normal no eixo traseiro $N_t$ se anule.

Assim, para encontrar a força $W$ necessária para provocar o tombamento, basta realizar uma somatória de momentos em torno do ponto $H$. Adotando o sentido anti-horário como positivo:

$$
\sum M_H = (2,9\,\mathrm{m})W+(0,9\,\mathrm{m})(3.000\,\mathrm{N})-(2,0\,\mathrm{m})(50.000\,\mathrm{N}) = 0
$$

A somatória dos momentos deve ser nula pois calcula-se a condição de início do tombamento. Nesta condição a máquina ainda se encontra na posição original e em repouso.



```python
W = (2*5e4-3000*0.9)/2.9

print(f'A força necessária é de W = {W:.2f} N')
```

Assim, para iniciar o tombamento se requer $W\geq33.551,72\,\mathrm{N}$. O valor mínimo de $W$ renderá a máxima aceleração que se pode aplicar na tora.

Observando apenas a tora, ela é puxada com a força $W$ para cima, pela empilhadeira, e com a força peso $P$ para baixo devido à atração gravitacional.

Sabe-se que o peso da tora é de $25\,\mathrm{kN}$, portanto sua massa é dada por:

$$
m = \frac{25.000,00\,\mathrm{N}}{9,81\,\mathrm{m/s²}} = 2.548,42\,\mathrm{kg}
$$

```python
m = 25000/9.81

print(f'A massa da tora é de m = {m:.2f} kg')
```

Assim, pode-se analisar a cinemática vertical da tora através da equação:

$$
ma = W-P ⇒a = \frac{W-P}{m} = \frac{33.551,72-25.000,00}{2.548,42} = 3,36 \,\mathrm{m/s^2}
$$

```python
a = (W-25000)/m

print(f'A máxima aceleração da tora é de a = {a:.2f} m/s²')
```

Ou seja, a tora pode sofrer uma aceleração máxima de $3,36\,\mathrm{m/s^2}$ para cima antes de provocar o tombamento da máquina.
